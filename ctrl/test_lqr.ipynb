```python

import numpy as np
from ctrl.control_system import DiscreteLTISystem


# define plant
dt = 1
A = np.array([[1, 1], [0, 1]])
B = np.array([[0], [1]])
C = np.eye(2)

plant = DiscreteLTISystem(A, B, C, dt=dt)
```

```python
from ctrl.controllers.lqr import LQRController
from ctrl.simulation import DiscreteControllerPlantObserverLoop

x0 = [10, 0]

Q = np.eye(2)
R = np.eye(1)

controller = LQRController(plant, Q, R)
sim = DiscreteControllerPlantObserverLoop(controller, plant)

x, _, _, _ = sim.sim(x0=x0, t=10)

import matplotlib.pyplot as plt
plt.plot(x)
```

```python
controller = LQRController(plant, Q, R, x_ref=[-5, 0])
sim = DiscreteControllerPlantObserverLoop(controller, plant)

x, _, _, _ = sim.sim(x0=x0, t=10)

import matplotlib.pyplot as plt
plt.plot(x)
```
