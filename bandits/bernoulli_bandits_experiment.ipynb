```python
from bandits_framework import BernoulliBanditsExperiment
from algorithms import UCB1BanditsAlgorithm

# experiment parameters
bandits_num = 10
T = 1000

alg = UCB1BanditsAlgorithm(bandits_num)
experiment = BernoulliBanditsExperiment(bandits_num, alg)

total_reward = experiment.run(T)
trajectory = experiment.trajectory
```

```python
import matplotlib.pyplot as plt

bandits_expected_reward = [params['p'] for params in experiment.bandits_params]

plt.figure()
plt.title('expected reward')
plt.bar(x=list(range(bandits_num)), height=bandits_expected_reward)
plt.xticks(range(bandits_num))
plt.xlabel('action')

plt.figure()
plt.title('trajectory')
plt.plot(trajectory, '.')
plt.ylabel('action')
plt.xlabel('t')

plt.figure()
plt.title('frequency of actions')
plt.hist(trajectory, bins=list(range(bandits_num + 1)), align='left')
plt.xticks(range(bandits_num))
```
